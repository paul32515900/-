```python
#Ë©¶Âà©Áî®ÂèçËΩâÊèõÊ≥ïÁî¢ÁîüLogarithmic(ùúÉ)ÁöÑÈö®Ê©üÊ®£Êú¨

import numpy as np

def rlogarithmic(n, theta):
    # returns a random logarithmic(theta) sample of size n
    u = np.random.uniform(size=n)  # generate n uniform random numbers

    # Set the initial length of cdf vector
    N = int(np.ceil(-16 / np.log10(theta)))
    k = np.arange(1, N + 1)
    a = -1 / np.log(1 - theta)
    
    # Calculate fk and cumulative distribution function (CDF)
    fk = np.exp(np.log(a) + k * np.log(theta) - np.log(k))
    Fk = np.cumsum(fk)
    
    # Initialize the output vector
    x = np.zeros(n, dtype=int)

    for i in range(n):
        x[i] = np.sum(u[i] > Fk)  # F^{-1}(u) - 1
        
        # If x == N, we need to extend the cdf
        while x[i] == N:
            logf = np.log(a) + (N + 1) * np.log(theta) - np.log(N + 1)
            fk = np.append(fk, np.exp(logf))  # extend fk
            Fk = np.append(Fk, Fk[N - 1] + fk[N])  # extend Fk
            N += 1
            x[i] = np.sum(u[i] > Fk)

    return x

#np.random.uniform(size=n) Áî®‰æÜÁîüÊàê n ÂÄã [0,1] ÂçÄÈñìÁöÑÂùáÂãªÈö®Ê©üÊï∏„ÄÇ
#np.arange(1, N + 1) ÁîüÊàê 1 Âà∞ N ÁöÑÂ∫èÂàóÂ∞çÊáâ R ‰∏≠ÁöÑ 1:N„ÄÇ
#np.cumsum() Âèñ‰ª£‰∫Ü R ‰∏≠ÁöÑ cumsum()ÔºåÁî®ÊñºË®àÁÆóÁ¥ØÁ©çÂàÜ‰ΩàÂáΩÊï∏„ÄÇ
#np.append() Áî®‰æÜÂú® fk Âíå Fk ‰∏≠Ê∑ªÂä†Êñ∞ÂÖÉÁ¥†ÔºåÊì¥Â±ïÈö®Ê©üÂàÜ‰Ωà„ÄÇ
#Ëø¥ÂúàÂÖßÁöÑ np.sum(u[i] > Fk) Áî®‰æÜÁ¢∫ÂÆöÈö®Ê©üÊï∏ÊáâË©≤Â±¨ÊñºÂì™ÂÄãÁØÑÂúçÁöÑÁ¥ØÁ©çÂàÜ‰Ωà„ÄÇ

```

```python
#Ë®àÁÆóùëôùëúùëîùëéùëüùëñùë°‚Ñéùëöùëñùëê(ùë°‚Ñéùëíùë°ùëé) ÁöÑÊ©üÁéáÂØÜÂ∫¶ÂáΩÊï∏‰ª•ÈÄ≤Ë°åÊØîËºÉ

import numpy as np
import pandas as pd

# ÂÅáË®≠Ê®£Êú¨ x, ÂèÉÊï∏ theta ÂíåÊ®£Êú¨Â§ßÂ∞è n Â∑≤Á∂ìÂÆöÁæ©
theta = 0.5  # ‰æãÂ¶ÇÔºåtheta ÁöÑÂÄº
x = np.array([0, 0, 0, 0, 1, 0, 1, 0, 0, 3])  # ÂÅáË®≠ÁöÑÈö®Ê©üÊ®£Êú¨
n = len(x)

# Ë®àÁÆóÂîØ‰∏ÄÁöÑ x ÂÄº‰∏¶ÊéíÂ∫èÔºåÂøΩÁï• k = 0 ÁöÑÊÉÖÊ≥Å
k = np.sort(np.unique(x[x > 0]))

# Ë®àÁÆóÂ∞çÊï∏ÂàÜÂ∏ÉÁöÑÊ©üÁéáÂØÜÂ∫¶ÂáΩÊï∏ (PDF)
p = -1 / np.log(1 - theta) * theta**k / k

# Ë®àÁÆóÊØèÂÄãÂÄºÁöÑÊ®ôÊ∫ñË™§Â∑Æ (SE)
se = np.sqrt(p * (1 - p) / n)

# Ë®àÁÆóÊ®£Êú¨ÁöÑÁõ∏Â∞çÈ†ªÁéá
freq_table = pd.value_counts(x, normalize=True).sort_index()

# Âè™‰øùÁïô k > 0 ÁöÑÈ†ªÁéá
freq_table = freq_table[freq_table.index > 0]

# Â∞áÁõ∏Â∞çÈ†ªÁéá„ÄÅÁêÜË´ñÊ©üÁéáÂíåÊ®ôÊ∫ñË™§Â∑ÆÂêà‰Ωµ
comparison_table = pd.DataFrame({
    'Relative Frequency': freq_table,
    'Theoretical Probability': p,
    'Standard Error': se
}).round(3)

comparison_table
```

```python
#Âà©Áî®Ê£ÑÂçªÊ≥ïÁî¢Áîü1000ÂÄãÊúçÂæûùêµùëíùë°ùëé(ùõº=2,ùõΩ =2)ÁöÑÈö®Ê©üÊ®£Êú¨
# Âëà‰∏äÈ°åÔºåÂà©Áî®Ê£ÑÂçªÊ≥ïÔºåÂπ≥Âùá‰æÜË™™ÈúÄË¶ÅÂ§öÂ∞ëÂÄãÈö®Ê©üÊï∏‰ª•Áî¢Áîü1000ÂÄãÊúçÂæûbùëíùë°ùëé(ùõº = 2,ùõΩ = 2)ÁöÑÈö®Ê©üÊ®£Êú¨
import numpy as np
import matplotlib.pyplot as plt

# Beta(2, 2) ÁöÑÊ¶ÇÁéáÂØÜÂ∫¶ÂáΩÊï∏
def beta_pdf(x):
    return 6 * x * (1 - x)

# ‰ΩøÁî®Ê£ÑÂçªÊ≥ïÁîüÊàê Beta(2, 2) Èö®Ê©üÊ®£Êú¨
def rejection_sampling_beta(n_samples):
    accepted_samples = []
    n_trials = 0  # Á∏ΩÂÖ±ÂòóË©¶ÁöÑÊ¨°Êï∏

    while len(accepted_samples) < n_samples:
        # ÁîüÊàêÂùáÂãªÂàÜ‰ΩàÁöÑÂÄôÈÅ∏Ê®£Êú¨
        x_candidate = np.random.uniform(0, 1)
        y_candidate = np.random.uniform(0, 6)  # 6 ÊòØÊàëÂÄëË®≠ÂÆöÁöÑ c ÂÄº
        
        n_trials += 1
        
        # Â¶ÇÊûúÂÄôÈÅ∏Ê®£Êú¨Ë¢´Êé•ÂèóÔºåÂâá‰øùÂ≠òÂÆÉ
        if y_candidate <= beta_pdf(x_candidate):
            accepted_samples.append(x_candidate)
    
    return np.array(accepted_samples), n_trials

# Áî¢Áîü 1000 ÂÄã Beta(2, 2) Èö®Ê©üÊ®£Êú¨
n_samples = 1000
samples, total_trials = rejection_sampling_beta(n_samples)

# Ë®àÁÆóÂπ≥ÂùáÁîüÊàêÈö®Ê©üÊï∏ÁöÑÊ¨°Êï∏
average_trials_per_sample = total_trials / n_samples
total_random_numbers = total_trials * 2  # ÊØèÊ¨°ÁîüÊàêÂÖ©ÂÄãÈö®Ê©üÊï∏ (x Âíå y)

# È°ØÁ§∫ÁµêÊûú
print(f"Âπ≥Âùá‰æÜË™™ÔºåÁîüÊàê {n_samples} ÂÄã Beta(2, 2) Èö®Ê©üÊ®£Êú¨ÈúÄË¶ÅÁöÑÈö®Ê©üÊï∏ÁÇ∫Ôºö{total_random_numbers}")

# Áï´Âá∫ÁîüÊàêÁöÑÊ®£Êú¨ÂàÜ‰Ωà
plt.hist(samples, bins=30, density=True, alpha=0.7, label="Generated Samples")
x = np.linspace(0, 1, 1000)
plt.plot(x, beta_pdf(x), 'r-', lw=2, label="Beta(2,2) PDF")
plt.title("Histogram of Generated Beta(2, 2) Samples")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#ÈÅãÁî®ËΩâÊèõÊ≥ïÁî¢Áîü‰∏Äùêµùëíùë°ùëé(3,2)ÁöÑÊ®£Êú¨
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# ÁîüÊàê Beta(3, 2) ÁöÑÊ®£Êú¨
def generate_beta_samples(alpha, beta_param, n_samples):
    # ÁîüÊàêÂùáÂãªÂàÜ‰ΩàÁöÑÈö®Ê©üÊï∏
    u = np.random.uniform(0, 1, n_samples)
    
    # ‰ΩøÁî®ÂèçÂáΩÊï∏Â∞áÂùáÂãªÈö®Ê©üÊï∏ËΩâÊèõÁÇ∫ Beta ÂàÜ‰ΩàÊ®£Êú¨
    samples = beta.ppf(u, alpha, beta_param)
    
    return samples

# ÂèÉÊï∏Ë®≠ÁΩÆ
alpha = 3
beta_param = 2
n_samples = 1000

# ÁîüÊàêÊ®£Êú¨
samples = generate_beta_samples(alpha, beta_param, n_samples)

# Áï´Âá∫ÁîüÊàêÁöÑÊ®£Êú¨ÂàÜ‰Ωà
plt.hist(samples, bins=30, density=True, alpha=0.7, label="Generated Samples")
x = np.linspace(0, 1, 1000)
plt.plot(x, beta.pdf(x, alpha, beta_param), 'r-', lw=2, label="Beta(3,2) PDF")
plt.title("Histogram of Generated Beta(3, 2) Samples")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
print(samples)
```

```python
import numpy as np

def rlogarithmic(n, theta):
    # Á¢∫‰øù theta Âú® (0, 1) ÁØÑÂúçÂÖß
    if not (0 < theta < 1):
        raise ValueError("theta must be in the range (0, 1)")

    # ÁîüÊàê n ÂÄãÂùáÂãªÂàÜ‰ΩàÁöÑÈö®Ê©üÊï∏ u Âíå v
    u = np.random.uniform(0, 1, n)
    v = np.random.uniform(0, 1, n)

    # Ë®àÁÆó x
    x = np.floor(1 + np.log(v) / np.log(1 - (1 - theta) ** u))

    return x.astype(int)  # ËøîÂõûÊï¥Êï∏È°ûÂûãÁöÑÁµêÊûú

# ‰ΩøÁî®ÁØÑ‰æã
n = 1000  # Èö®Ê©üÊ®£Êú¨ÁöÑÂ§ßÂ∞è
theta = 0.5  # ÂèÉÊï∏ theta
sample = rlogarithmic(n, theta)

# È°ØÁ§∫ÁîüÊàêÁöÑÊ®£Êú¨
print("ÁîüÊàêÁöÑÂ∞çÊï∏ÂàÜ‰ΩàÈö®Ê©üÊ®£Êú¨Ôºö")
print(sample)
```

Ë®≠ $X_1 \sim \text{Gamma}(2, 2)$ ‰∏î $X_2 \sim \text{Gamma}(2, 4)$ ÁÇ∫Áç®Á´ãÁöÑÈö®Ê©üËÆäÊï∏„ÄÇÊØîËºÉ‰ª•‰∏ãÂÖ©Á®ÆÈö®Ê©üÊ®£Êú¨ÁöÑÁõ¥ÊñπÂúñÔºö

1. **Êë∫Á©çÊ®£Êú¨**Ôºö$S = X_1 + X_2$ÔºåÂç≥ $S$ ÊòØ $X_1$ Âíå $X_2$ ÁöÑÊë∫Á©çÂàÜ‰Ωà„ÄÇ

2. **Ê∑∑ÂêàÂàÜ‰ΩàÊ®£Êú¨**Ôºö$F_X = 0.5F_{X_1} + 0.5F_{X_2}$ÔºåÂç≥ $F_X$ ÊòØ‰æÜËá™ $X_1$ Âíå $X_2$ ÂêÑ 50% Ê©üÁéáÊ∑∑ÂêàÂàÜ‰ΩàÁöÑÊ®£Êú¨„ÄÇ

```python
#Ë®≠ùëã1 ‚àº ùê∫ùëéùëöùëöùëé(2,2) ‰∏î ùëã2 ‚àº ùê∫ùëéùëöùëöùëé(2,4) ÁÇ∫Áç®Á´ãÁöÑÔºõÊØîËºÉÁî±Êë∫Á©ç S =ùëã1+ùëã2Áî¢ÁîüÁöÑÊ®£Êú¨ËàáÊ∑∑Âêàùêπùëã=0.5ùêπùëã1+0.5ùêπùëã2 Áî¢ÁîüÁöÑÊ®£Êú¨ÁöÑÁõ¥ÊñπÂúñ

import numpy as np
import matplotlib.pyplot as plt

# ÂÆöÁæ©Ê®£Êú¨Êï∏ n
n = 1000

# 1. ÁîüÊàêÂÖ©ÂÄã Gamma ÂàÜ‰ΩàÁöÑÊ®£Êú¨
x1 = np.random.gamma(2, 1/2, n)  # shape=2, rate=2 => scale = 1/rate = 1/2
x2 = np.random.gamma(2, 1/4, n)  # shape=2, rate=4 => scale = 1/rate = 1/4

# 2. Ë®àÁÆóÁ∏ΩÂíå
s = x1 + x2

# 3. ÁîüÊàêÂùáÂãªÂàÜ‰ΩàÁöÑÈö®Ê©üÊï∏
r = np.random.uniform(0, 1, n)

# 4. Ê†πÊìöÊ¢ù‰ª∂ÈÅ∏Êìá
k = (r > 0.5).astype(int)  # Â∞áÂ∏ÉÁàæÂÄºËΩâÊèõÁÇ∫ 0 Êàñ 1
x = k * x1 + (1 - k) * x2

# 5. Áπ™Ë£ΩÁõ¥ÊñπÂúñ
plt.figure(figsize=(10, 5))

# Á¨¨‰∏ÄÂÄãÂúñÔºös ÁöÑÁõ¥ÊñπÂúñ
plt.subplot(1, 2, 1)
plt.hist(s, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Histogram of s')

# Á¨¨‰∫åÂÄãÂúñÔºöx ÁöÑÁõ¥ÊñπÂúñ
plt.subplot(1, 2, 2)
plt.hist(x, bins=30, density=True, alpha=0.7, color='green')
plt.title('Histogram of x')

# È°ØÁ§∫ÂúñÂΩ¢
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ë®≠ÂÆöÂèÉÊï∏
n = 5000
k = np.random.choice(np.arange(1, 6), size=n, replace=True, p=(np.arange(1, 6) / 15))
rate = 1 / k

# ÁîüÊàê gamma ÂàÜ‰ΩàÈö®Ê©üËÆäÊï∏
x = np.random.gamma(shape=3, scale=1/rate)

# Áπ™Ë£ΩÊ∑∑ÂêàÂàÜ‰ΩàÁöÑÂØÜÂ∫¶
plt.figure(figsize=(10, 6))
sns.kdeplot(x, lw=3, label='Mixture Density')

# Áπ™Ë£ΩÁµÑÊàêÈÉ®ÂàÜÁöÑÂØÜÂ∫¶
for i in range(1, 6):
    component_samples = np.random.gamma(shape=3, scale=1/i, size=n)
    sns.kdeplot(component_samples, lw=1, label=f'Component {i}')

# Ë®≠ÂÆö x Ëª∏Âíå y Ëª∏ÁöÑÁØÑÂúç
plt.xlim(0, 40)
plt.ylim(0, 0.3)

# Ê∑ªÂä†Ê®ôÁ±§ÂíåÂúñ‰æã
plt.xlabel('x')
plt.title('Density of the Mixture with Component Densities')
plt.legend()
plt.show()

```

```python
import numpy as np


n = 5000
p = [0.1, 0.2, 0.2, 0.3, 0.2]
lambda_values = [1, 1.5, 2, 2.5, 3]

# Ê†πÊìöÊ©üÁéá pÊäΩÊ®£k
k = np.random.choice(np.arange(1, 6), size=n, replace=True, p=p)

# Ê†πÊìö k Áç≤ÂèñÂ∞çÊáâÁöÑrate
rate = np.array([lambda_values[i-1] for i in k])


x = np.random.gamma(shape=3, scale=1/rate)
print(x)


```

Áπ™Ë£Ω Gamma ÂàÜÈÖçÊ©üÁéáÂØÜÂ∫¶ÂèäÊ∑∑ÂêàÂàÜÈÖçÁöÑÊ©üÁéáÂØÜÂ∫¶

Ê∑∑ÂêàÂàÜÈÖçÁöÑÊ©üÁéáÂØÜÂ∫¶ÂáΩÊï∏ÁÇ∫Ôºö

$$
f_X(x) = \sum_{j=1}^{5} \theta_j f_j(x), \quad x > 0
$$

ÂÖ∂‰∏≠Ôºå$( X_j \sim \text{Gamma}(r = 3, \lambda_j) $) ÁÇ∫Áç®Á´ãÁöÑ


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# ÂÆöÁæ©Ê∑∑ÂêàÂàÜÈÖçÁöÑÂØÜÂ∫¶ÂáΩÊï∏
def f(x, lambda_values, theta):
    return np.sum([gamma.pdf(x, a=3, scale=1/lambda_values[i]) * theta[i] for i in range(len(lambda_values))])

# ÂèÉÊï∏
p = [0.1, 0.2, 0.2, 0.3, 0.2]
lambda_values = [1, 1.5, 2, 2.5, 3]
x = np.linspace(0, 8, 200)

# Ë®àÁÆóÊ∑∑ÂêàÂàÜÈÖçÁöÑÂØÜÂ∫¶ÂáΩÊï∏ÂÄº
y_mixture = np.array([f(val, lambda_values, p) for val in x])

# Áπ™Ë£ΩÊ∑∑ÂêàÂàÜÈÖçÁöÑÂØÜÂ∫¶Âúñ
plt.plot(x, y_mixture, label='Mixture Density', lw=3)
plt.ylim(0, 0.85)
plt.ylabel("Density")
plt.title("Gamma and Mixture Densities")

# Áπ™Ë£ΩÂêÑÂÄã Gamma ÂàÜÈÖçÁöÑÂØÜÂ∫¶Âúñ
for j in range(5):
    y_gamma = gamma.pdf(x, a=3, scale=1/lambda_values[j])
    plt.plot(x, y_gamma, label=f'Gamma density Œª={lambda_values[j]}')

# È°ØÁ§∫Âúñ‰æã
plt.legend()
plt.show()
```

## Poisson-Gamma Ê∑∑ÂêàÂàÜÈÖçËàáË≤†‰∫åÈ†ÖÂàÜÈÖç

Ë≤†‰∫åÈ†ÖÂàÜÈÖçÂèØ‰ª•Ë¢´Ë¶ñÁÇ∫ Poisson ÂàÜÈÖçÁöÑÊ∑∑ÂêàÂàÜÈÖçÔºåÂÖ∂‰∏≠ $(\Lambda$) ÊòØ‰∏ÄÂÄã Gamma ÂàÜÈÖç„ÄÇ

Ëã•Ôºö

- $((X | \Lambda) \sim \text{Poisson}(\Lambda)$)
- $(\Lambda \sim \text{Gamma}(r, \beta)$)

Ââá $(X$) ÊúçÂæûË≤†‰∫åÈ†ÖÂàÜÈÖçÔºåÂÖ∂ÂèÉÊï∏ÁÇ∫ $(r$) Âèä $(p = \frac{\beta}{1 + \beta}$)„ÄÇ

### Ë©¶Âà©Áî®‰∏äËø∞Èóú‰øÇÁî¢Áîü Poisson-Gamma Ê∑∑ÂêàÈö®Ê©üËÆäÊï∏Ôºå‰∏¶ËàáË≤†‰∫åÈ†ÖÂàÜÈÖçÁöÑÈö®Ê©üËÆäÊï∏ÈÄ≤Ë°åÊØîËºÉ„ÄÇ

```python
import numpy as np
import pandas as pd
from scipy.stats import gamma, poisson, nbinom

# ÂèÉÊï∏Ë®≠ÁΩÆ
n = 1000  # ÁîüÊàêÁöÑÊ®£Êú¨Êï∏
r = 4     # Ë≤†‰∫åÈ†ÖÂàÜÈÖçÁöÑ r ÂèÉÊï∏
beta = 3  # Gamma ÂàÜÈÖçÁöÑ beta ÂèÉÊï∏

# ÁîüÊàê Gamma Èö®Ê©üËÆäÊï∏‰ΩúÁÇ∫ Poisson ÁöÑÂùáÂÄº
lambda_samples = gamma.rvs(a=r, scale=1/beta, size=n)

# ‰ΩøÁî®ÁîüÊàêÁöÑ lambda ÁîüÊàê Poisson Èö®Ê©üËÆäÊï∏
x = poisson.rvs(lambda_samples)

# Ë®àÁÆóÊ∑∑ÂêàÂàÜÈÖçÁöÑÈ†ªÁéá
mix = np.bincount(x) / n

# Ë®àÁÆóË≤†‰∫åÈ†ÖÂàÜÈÖçÁöÑÊ¶ÇÁéá
max_x = np.max(x)
negbin = nbinom.pmf(range(max_x + 1), r, beta / (1 + beta))

# Ë®àÁÆóÊ®ôÊ∫ñË™§Â∑Æ
se = np.sqrt(negbin * (1 - negbin) / n)

# Â∞áÁµêÊûúÊï¥ÁêÜÊàê DataFrame Êñπ‰æøÈ°ØÁ§∫
result = pd.DataFrame({
    'Mixture': np.round(mix, 3),
    'Negative Binomial': np.round(negbin, 3),
    'SE': np.round(se, 3)
}).fillna(0)  # Áî® 0 Â°´ÂÖÖÁ©∫ÂÄº

print(result)
```
