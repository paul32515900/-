```python
#試利用反轉換法產生Logarithmic(𝜃)的隨機樣本

import numpy as np

def rlogarithmic(n, theta):
    # returns a random logarithmic(theta) sample of size n
    u = np.random.uniform(size=n)  # generate n uniform random numbers

    # Set the initial length of cdf vector
    N = int(np.ceil(-16 / np.log10(theta)))
    k = np.arange(1, N + 1)
    a = -1 / np.log(1 - theta)
    
    # Calculate fk and cumulative distribution function (CDF)
    fk = np.exp(np.log(a) + k * np.log(theta) - np.log(k))
    Fk = np.cumsum(fk)
    
    # Initialize the output vector
    x = np.zeros(n, dtype=int)

    for i in range(n):
        x[i] = np.sum(u[i] > Fk)  # F^{-1}(u) - 1
        
        # If x == N, we need to extend the cdf
        while x[i] == N:
            logf = np.log(a) + (N + 1) * np.log(theta) - np.log(N + 1)
            fk = np.append(fk, np.exp(logf))  # extend fk
            Fk = np.append(Fk, Fk[N - 1] + fk[N])  # extend Fk
            N += 1
            x[i] = np.sum(u[i] > Fk)

    return x

#np.random.uniform(size=n) 用來生成 n 個 [0,1] 區間的均勻隨機數。
#np.arange(1, N + 1) 生成 1 到 N 的序列對應 R 中的 1:N。
#np.cumsum() 取代了 R 中的 cumsum()，用於計算累積分佈函數。
#np.append() 用來在 fk 和 Fk 中添加新元素，擴展隨機分佈。
#迴圈內的 np.sum(u[i] > Fk) 用來確定隨機數應該屬於哪個範圍的累積分佈。

```

```python
#計算𝑙𝑜𝑔𝑎𝑟𝑖𝑡ℎ𝑚𝑖𝑐(𝑡ℎ𝑒𝑡𝑎) 的機率密度函數以進行比較

import numpy as np
import pandas as pd

# 假設樣本 x, 參數 theta 和樣本大小 n 已經定義
theta = 0.5  # 例如，theta 的值
x = np.array([0, 0, 0, 0, 1, 0, 1, 0, 0, 3])  # 假設的隨機樣本
n = len(x)

# 計算唯一的 x 值並排序，忽略 k = 0 的情況
k = np.sort(np.unique(x[x > 0]))

# 計算對數分布的機率密度函數 (PDF)
p = -1 / np.log(1 - theta) * theta**k / k

# 計算每個值的標準誤差 (SE)
se = np.sqrt(p * (1 - p) / n)

# 計算樣本的相對頻率
freq_table = pd.value_counts(x, normalize=True).sort_index()

# 只保留 k > 0 的頻率
freq_table = freq_table[freq_table.index > 0]

# 將相對頻率、理論機率和標準誤差合併
comparison_table = pd.DataFrame({
    'Relative Frequency': freq_table,
    'Theoretical Probability': p,
    'Standard Error': se
}).round(3)

comparison_table
```

```python
#利用棄卻法產生1000個服從𝐵𝑒𝑡𝑎(𝛼=2,𝛽 =2)的隨機樣本
# 呈上題，利用棄卻法，平均來說需要多少個隨機數以產生1000個服從b𝑒𝑡𝑎(𝛼 = 2,𝛽 = 2)的隨機樣本
import numpy as np
import matplotlib.pyplot as plt

# Beta(2, 2) 的概率密度函數
def beta_pdf(x):
    return 6 * x * (1 - x)

# 使用棄卻法生成 Beta(2, 2) 隨機樣本
def rejection_sampling_beta(n_samples):
    accepted_samples = []
    n_trials = 0  # 總共嘗試的次數

    while len(accepted_samples) < n_samples:
        # 生成均勻分佈的候選樣本
        x_candidate = np.random.uniform(0, 1)
        y_candidate = np.random.uniform(0, 6)  # 6 是我們設定的 c 值
        
        n_trials += 1
        
        # 如果候選樣本被接受，則保存它
        if y_candidate <= beta_pdf(x_candidate):
            accepted_samples.append(x_candidate)
    
    return np.array(accepted_samples), n_trials

# 產生 1000 個 Beta(2, 2) 隨機樣本
n_samples = 1000
samples, total_trials = rejection_sampling_beta(n_samples)

# 計算平均生成隨機數的次數
average_trials_per_sample = total_trials / n_samples
total_random_numbers = total_trials * 2  # 每次生成兩個隨機數 (x 和 y)

# 顯示結果
print(f"平均來說，生成 {n_samples} 個 Beta(2, 2) 隨機樣本需要的隨機數為：{total_random_numbers}")

# 畫出生成的樣本分佈
plt.hist(samples, bins=30, density=True, alpha=0.7, label="Generated Samples")
x = np.linspace(0, 1, 1000)
plt.plot(x, beta_pdf(x), 'r-', lw=2, label="Beta(2,2) PDF")
plt.title("Histogram of Generated Beta(2, 2) Samples")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#運用轉換法產生一𝐵𝑒𝑡𝑎(3,2)的樣本
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# 生成 Beta(3, 2) 的樣本
def generate_beta_samples(alpha, beta_param, n_samples):
    # 生成均勻分佈的隨機數
    u = np.random.uniform(0, 1, n_samples)
    
    # 使用反函數將均勻隨機數轉換為 Beta 分佈樣本
    samples = beta.ppf(u, alpha, beta_param)
    
    return samples

# 參數設置
alpha = 3
beta_param = 2
n_samples = 1000

# 生成樣本
samples = generate_beta_samples(alpha, beta_param, n_samples)

# 畫出生成的樣本分佈
plt.hist(samples, bins=30, density=True, alpha=0.7, label="Generated Samples")
x = np.linspace(0, 1, 1000)
plt.plot(x, beta.pdf(x, alpha, beta_param), 'r-', lw=2, label="Beta(3,2) PDF")
plt.title("Histogram of Generated Beta(3, 2) Samples")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
print(samples)
```

設 $X$ 的混合分配為：

$$
X = \sum_{i=1}^{5} \theta_i F_{X_i}
$$

其中，$X_j \sim \text{Gamma}(r=3, \lambda_j = \frac{1}{j})$ 為獨立的，且混合的機率為 $\theta_j = \frac{j}{5}, j = 1, 2, \dots, 5$


### 解

為了從混合分配 $F_X$ 中模擬出一個樣本，步驟如下：

1. 產生一個整數 $k \in \{1, 2, 3, 4, 5\}$，其中 $P(k) = \theta_k$，且 $k = 1, \dots, 5$。
   
2. 輸出一個服從 $\text{Gamma}(r, \lambda_k)$ 的隨機變數。



若欲產生一組大小為 $n$ 的隨機樣本，上述步驟需重複 $n$ 次。注意在 R 中，`for` 迴圈效率不高，因此我們可以使用向量化的方法來產生此組樣本。

#### 步驟：

1. 產生 $k$ 的隨機樣本 $\{k_1, k_2, \dots, k_n\}$，其中 $P(k) = \theta_k$，$k = 1, \dots, 5$。這樣，$k[i]$ 表示第 $i$ 個樣本將從這五個 Gamma 分配中的哪一個中抽出。

2. 設定 `rate` 參數為長度為 $n$ 的向量 $\lambda = (\lambda_k)$。

3. 產生一組大小為 $n$ 的 Gamma 分配隨機樣本，其形狀參數為 $r$，`rate` 參數為 `rate` 向量中的對應值。

繪製 Gamma 分配機率密度及混合分配的機率密度

混合分配的機率密度函數為：

$$
f_X(x) = \sum_{j=1}^{5} \theta_j f_j(x), \quad x > 0
$$

其中，$( X_j \sim \text{Gamma}(r = 3, \lambda_j) $) 為獨立的


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# 定義混合分配的密度函數
def f(x, lambda_values, theta):
    return np.sum([gamma.pdf(x, a=3, scale=1/lambda_values[i]) * theta[i] for i in range(len(lambda_values))])

# 參數
p = [0.1, 0.2, 0.2, 0.3, 0.2]
lambda_values = [1, 1.5, 2, 2.5, 3]
x = np.linspace(0, 8, 200)

# 計算混合分配的密度函數值
y_mixture = np.array([f(val, lambda_values, p) for val in x])

# 繪製混合分配的密度圖
plt.plot(x, y_mixture, label='Mixture Density', lw=3)
plt.ylim(0, 0.85)
plt.ylabel("Density")
plt.title("Gamma and Mixture Densities")

# 繪製各個 Gamma 分配的密度圖
for j in range(5):
    y_gamma = gamma.pdf(x, a=3, scale=1/lambda_values[j])
    plt.plot(x, y_gamma, label=f'Gamma density λ={lambda_values[j]}')

# 顯示圖例
plt.legend()
plt.show()
```

## Poisson-Gamma 混合分配與負二項分配

負二項分配可以被視為 Poisson 分配的混合分配，其中 $(\Lambda$) 是一個 Gamma 分配。

若：

- $((X | \Lambda) \sim \text{Poisson}(\Lambda)$)
- $(\Lambda \sim \text{Gamma}(r, \beta)$)

則 $(X$) 服從負二項分配，其參數為 $(r$) 及 $(p = \frac{\beta}{1 + \beta}$)。

### 試利用上述關係產生 Poisson-Gamma 混合隨機變數，並與負二項分配的隨機變數進行比較。

```python
import numpy as np
import pandas as pd
from scipy.stats import gamma, poisson, nbinom

# 參數設置
n = 1000  # 生成的樣本數
r = 4     # 負二項分配的 r 參數
beta = 3  # Gamma 分配的 beta 參數

# 生成 Gamma 隨機變數作為 Poisson 的均值
lambda_samples = gamma.rvs(a=r, scale=1/beta, size=n)

# 使用生成的 lambda 生成 Poisson 隨機變數
x = poisson.rvs(lambda_samples)

# 計算混合分配的頻率
mix = np.bincount(x) / n

# 計算負二項分配的概率
max_x = np.max(x)
negbin = nbinom.pmf(range(max_x + 1), r, beta / (1 + beta))

# 計算標準誤差
se = np.sqrt(negbin * (1 - negbin) / n)

# 將結果整理成 DataFrame 方便顯示
result = pd.DataFrame({
    'Mixture': np.round(mix, 3),
    'Negative Binomial': np.round(negbin, 3),
    'SE': np.round(se, 3)
}).fillna(0)  # 用 0 填充空值

print(result)
```
